```python
from sklearn.datasets import load_iris
from sklearn import tree
import pandas as pd
import numpy as np
url = "iris.data"  
names = ['花萼-length', '花萼-width', '花瓣-length', '花瓣-width', 'class']  
dataset = pd.read_csv(url, names=names)
dataset
```

```python
X,y=dataset.iloc[:,:4],dataset.iloc[:,4]
```

```python
X.shape,y.shape
```

```python
clf = tree.DecisionTreeClassifier(criterion='entropy')
clf = clf.fit(X, y)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(12,12))
tree.plot_tree(clf,fontsize=14)
```

```python
from sklearn.tree import export_text
r=export_text(clf)
print(r)
```

```python
data_file_encode = "gb18030"  # the watermelon_3.csv is file codec type
with open("4.3_ID3/data/watermelon_3.csv", mode = 'r', encoding = data_file_encode) as data_file:
    df = pd.read_csv(data_file)
df
```

```python
X,y=df.iloc[:,1:9],df.iloc[:,9]
```

```python
from sklearn import preprocessing
enc=preprocessing.OneHotEncoder()
e=enc.fit_transform(X.iloc[:,:6]).toarray()
```

```python
e[:10,:3]
```

```python
X=np.concatenate([e,X.iloc[:,6:]],axis=1)
X[:5,:]
```

```python
clf = tree.DecisionTreeClassifier()
clf = clf.fit(X, y)
```

```python
# 也可以用决策树进行回归
rg=tree.DecisionTreeRegressor()
rg=rg.fit(X[:,:6],X[:,6])
```

```python
from sklearn.metrics import mean_squared_error
y_pred=rg.predict(X[:,:6])
mean_squared_error(y_pred,X[:,6])
```
